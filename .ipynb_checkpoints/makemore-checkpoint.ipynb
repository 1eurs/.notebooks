```python
words = open('names.txt','r').read().splitlines()
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}

# b= {}
# for w in words:
#     chs = ['.'] + list(w) + ['']
#     for ch1,ch2 in zip(chs,chs[1:]):
#         bigram = (ch1,ch2)
#         b[bigram] = b.get(bigram,0)+1

# bsorted = sorted(b.items(), key=lambda x: x[1], reverse=True)
```

```python
import torch

N = torch.zeros((27,27), dtype=torch.int32)

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs,chs[1:]):
        x1 = stoi[ch1]
        x2 = stoi[ch2]
        N[x1,x2]+=1
    
```

```python
# import seaborn as sns

# plt.figure(figsize=(16,16))
# sns.heatmap(P, cmap="rocket")
# plt.show()

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(27):
    for j in range(27):
        chstr = itos[i] + itos[j]
        plt.text(i,j,chstr, ha='center' , va='bottom', color='gray')
        plt.text(j,i,N[i,j].item(),ha='center' , va='top', color='gray')
```

```python
P = N.float()
P /=  P.sum(1, keepdim=True)
```

```python
g = torch.Generator().manual_seed(2224411444222224222)

for _ in range(5): 
    w = []
    ix = 0
    while True:
        p = P[ix] + 1
        # p = N[ix].float()
        # p = p / p.sum()
        # p = torch.ones(27)/27.0
        ix = torch.multinomial(p ,1, replacement=True,generator=g).item()
        w.append(itos[ix])
        if ix == 0:
            break
            
    print(''.join(w))
```

```python
# a*b*c
# log(a*b*c) = log(a) + log(b) + log(c)
```

```python
log_likelihood = 0.0
n = 0
for w in ["andrejq"]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs,chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1,ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        print(f"{ch1}{ch2}: {prob:.4f} {logprob:.4f}")
        

print(f"{log_likelihood} ||| {(-log_likelihood)} ||| {(-log_likelihood)/n}")
```

```python
xs,ys = [], []

for w in words[:1]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs,chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        # print(f"{ch1} {ch2}")
        xs.append(ix1)
        ys.append(ix2)
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
import torch.nn.functional as F
xenc = F.one_hot(xs, 27).float()
plt.imshow(xenc)
```

```python
W = torch.randn((27,27))
```

```python
# (5),27 @ 27,(1) = 5,1
# 5,27 @ 27,27 = 5,27
logits = xenc @ W
counts = logits.exp()
prob = counts / counts.sum(1, keepdims = True)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3))
for ax, d, t in zip(axs, [xenc, W, probs], ['xenc', 'W', 'probs']):
    ax.imshow(d, cmap='viridis')
plt.tight_layout()
plt.show()
```
